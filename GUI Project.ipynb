```python
#importing the GUI library in python
from  tkinter import *
from datetime import *

# Building the windows


frame = Tk()
frame.title("Foster's work")
frame.geometry("450x500")
frame.config(bg = 'RED')



#Builing the labels
Label(frame,text= "Hello!!! It's Your Project Calculator ",bg ="white", font = ("Poppins", 17,"bold","underline"), ).pack(pady=10)
Label(frame,text= "Project Name:  ",bg ="white", font = ("Poppins", 12,"underline")).place(x=15,y=70)
Label(frame,text= "Start Date ",bg ="white", font = ("Poppins", 12,"underline")).place(x= 15,y= 120)
Label(frame,text= "Finish Date ",bg ="white", font = ("Poppins", 12,"underline")).place(x= 15,y= 170)
Label(frame,text= "Output",bg ="white", font = ("Poppins", 12,"underline")).place(x= 15,y= 340)


#Creating Entries
project_name = Entry(frame, text ="Project Name", font = ('Poppins', 15))
start_date = Entry(frame, text ="Start Date", font = ('Poppins', 15))
finish_date = Entry(frame, text ="Finish Date", font = ('Poppins', 15))



#positioning the boxes
project_name.place(x=200,y=70)
start_date.place(x=200,y=120)
finish_date.place(x=200,y=170)

new_list = []


#calculate

def calculate():

    global new_list
    project_name_1 = project_name.get()
    start_date_1 = start_date.get()
    finish_date_1 = finish_date.get()


    project_name_2 = project_name_1
    print(project_name_2)
    start_date_2 = datetime.strptime(start_date_1,"")
    finish_date_1 = datetime.strptime(finish_date_1,"" )
    differ = finish_date_1 - start_date_2
    number_of_days = differ.days
    days_to_hours = number_of_days * 24
    revenue = days_to_hours * 7
    final_output = "Revenue is $ " + str(revenue)

    Label(frame, text=final_output, bg="white", fg = 'red',font=("Poppins", 12, "italic")).place(x=200, y=340)









#clearing the text in the boxes
def clear():
    project_name.delete(0,END)
    start_date.delete(0,END)
    finish_date.delete(0,END)












#Submit_Button =
my_button =Button(frame,text ="Calculate",command=calculate(), font=("poppins", 14,"underline"),bg ="Pink").place(x =200,y =270)
mainloop()
```
